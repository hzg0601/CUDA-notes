### 任务3 张三丰的方圆之道
任务描述:
武当张三丰潜心研究太极拳,  在连续画圆几十年后, 成功领悟了物理/数学不分家的道理. 现在他决定用自己强悍的臂力, 以蒙特卡洛法, 在一个2001米 * 2001米的场地内进行真随机投掷7公斤重的铅球10W次. 以求得数学中的究极魔法数字Pi的近似值. 
Ken听闻此事后, 大为老人的精神感动, 决定带领第三届的同学们, 用GPU替他完成统计计算.
现在需要勇敢的你继续去完张三丰的未竟事业, 铅球的落点坐标, Ken老师已经为你整理成2D数组的形式.  

详细描述：利用蒙特卡洛法求取圆周率Pi值。在一个正方形内，真随机投掷一个小球，计算小球与正方形正中心的距离，如果小于圆的半径我们则认为球在圆内。那么就可以利用在圆内球的个数I 和在总的球的个数N求出圆周率Pi = （4\*I)/N  
具体代码请参考3.cu

任务难度: B
前置资源: 1x 未完成的代码框架, 1x公式: Pi = 落在大方形的内切圆中的铅球数量 / 2500. 
要求: 完成框架, 得到Pi的近似值(不能偏差1%以上).   
任务奖惩:  
成功奖励: NVIDIA的CUDA on ARM课程证书 的最后1/3部分
